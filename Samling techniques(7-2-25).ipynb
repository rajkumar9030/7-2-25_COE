```python
import pandas as pd 
from sklearn.utils import resample

df=pd.DataFrame({
    'age':[22,25,27,28,30,35,40,45,50,55,60,65,70],
    'income':[2000,2500,2700,3200,3500,3800,4000,4200,4300,4500,5000,5500,6000],
    'class':["high","low","low","high",'high','low','high','high','low','low','high','high','low']
})
df
```

```python
df_high=df[df['class']=='high']
df_low=df[df['class']=='low']
print(df_high)
print(df_low)
```

```python
df_high_downsampled=resample(df_high,replace=False,n_samples=len(df_low),random_state=42)
```

```python
df_balanced=pd.concat([df_high_downsampled,df_low])
```

```python
print(df_balanced['class'].value_counts())
df_balanced
```

```python
import pandas as pd 
from sklearn.utils import resample

df1=pd.DataFrame({
    'age':[22,25,27,28,30,35,40,45,50,55,60,65,70],
    'income':[2000,2500,2700,3200,3500,3800,4000,4200,4300,4500,5000,5500,6000],
    'class':["maj","min","min","maj",'maj','min','maj','maj','min','min','maj','maj','min']
})
print(df1)
print()
df_maj=df1[df1['class']=='maj']
df_min=df1[df1['class']=='min']
print(df_maj)
print(df_min)


df_min_upsampled=resample(df_min,replace=True,n_samples=len(df_maj),random_state=42)
df_balanced_2=pd.concat([df_min_upsampled,df_maj])
print(df_balanced_2['class'].value_counts())
df_balanced_2
```
